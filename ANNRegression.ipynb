```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
regression = Sequential()
```

```python
regression.add(Dense(units = 6 ,kernel_initializer="random_uniform",activation = "relu"))
```

```python
regression.add(Dense(units = 12 ,kernel_initializer="random_uniform",activation = "relu"))
```

```python
regression.add(Dense(units = 12 ,kernel_initializer="random_uniform",activation = "relu"))
```

```python
regression.add(Dense(units = 12 ,kernel_initializer="random_uniform",activation = "relu"))
```

```python
regression.add(Dense(units = 1 ,kernel_initializer="random_uniform",activation = "relu"))
```

```python
regression.compile(optimizer = "adam", loss = "mean_squared_error",metrics = ["accuracy"])
```

```python
regression.fit(x_train,y_train,epochs=500)
```

```python
ypred = regression.predict(x_test)
```
