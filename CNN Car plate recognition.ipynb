```python
#This demonstrates the recognition of car plate number using cnn
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Convolution2D,MaxPooling2D,Flatten
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_dataplate = ImageDataGenerator(rescale = 1./255,shear_range = 0.2,zoom_range = 0.2 , horizontal_flip = True)
test_dataplate = ImageDataGenerator(rescale = 1./255)
```

```python
#batch_size = 32
x_train = train_dataplate.flow_from_directory(r"C:\koe\train_set",target_size = (32,32),batch_size = 16, class_mode = "categorical")
x_test = test_dataplate.flow_from_directory(r"C:\koe\test_set",target_size = (32,32),batch_size = 16, class_mode = "categorical")
```

```python
cnnplate = Sequential()
```

```python
cnnplate.add(Convolution2D(16,(3,3),input_shape= (32,32,3)))
```

```python
cnnplate.add(MaxPooling2D((2,2)))
```

```python
cnnplate.add(Flatten()) # inputs to ann 
```

```python
cnnplate.add(Dense(units =64 ,kernel_initializer = "random_uniform",activation = "relu"))
```

```python
cnnplate.add(Dense(units = 6  ,kernel_initializer = "random_uniform",activation = "softmax"))
```

```python
cnnplate.compile(optimizer = "rmsprop",loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
cnnplate.fit_generator(x_train , steps_per_epoch = 54/16 , epochs = 200, validation_data = x_test , validation_steps = 26/16 )
```

```python
cnnplate.save("plate.h5")
```
